```python
import ROOT
import numpy as np 
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
f = ROOT.TFile('./allpart.root','r') 
tree = f.Get('randomParticles')
```

```python
df = ROOT.RDataFrame("randomParticles", "./allpart.root")
```

```python
secondary_type = df.AsNumpy(['fSecondaryType'])
coor = df.AsNumpy(['fX','fY'])
showerID = df.AsNumpy(['fPrimShowerID'])
primary_info = df.AsNumpy(['fPrimShowerID','fPrimParticle','fPrimEnergy','fPrimTheta'])
```

```python
type_sec = secondary_type['fSecondaryType']
ID = showerID['fPrimShowerID']
type_mc = primary_info['fPrimParticle']
e_mc = primary_info['fPrimEnergy']
theta_mc = primary_info['fPrimTheta']
```

```python
print('Number of simulated showers: ',len(ID))
```

```python
sec_id_hist = plt.hist(type_sec,bins=np.arange(0,15,1))
plt.yscale('log')
```

```python
plt.subplots(1,2,figsize=(15,5), dpi=500)

plt.subplot(1,2,1)
type_hist = plt.hist(np.log10(type_mc), np.arange(1,10,0.1))
plt.xlabel('Primary ID',size=14)
plt.yscale('log')

plt.subplot(1,2,2)
type_hist = plt.hist(np.log(e_mc), np.arange(0,15,1))
plt.xlabel('log ( Primary Energy / eV ) ',size=14)
```

```python
plt.subplots(1,3,figsize=(15,5), dpi=500)


plt.subplot(1,3,1)
type_hist = plt.hist(theta_mc, np.arange(0,90,1))
plt.xlabel(r'Primary $\Theta$ (deg)',size=14)

plt.subplot(1,3,2)
type_hist = plt.hist(np.cos(theta_mc*np.pi/180), np.arange(0,1,0.05))
plt.xlabel(r'Primary cos($\Theta$)',size=14)

plt.subplot(1,3,3)
type_hist = plt.hist(np.cos(theta_mc*np.pi/180)**2, np.arange(0,1,0.05))
plt.xlabel(r'Primary $\cos^2(\Theta)$',size=14)
```
